В скольких столбцах таблицы среднее значение измеряется в сотнях? (не в десятках и не в тысячах).

```python
numeric = df_train.select_dtypes(include=np.number)
count = 0
for column in numeric.columns:
  print(df_train[column].mean())
  if (df_train[column].mean() >= 100) and (df_train[column].mean() < 1000):
    count += 1
print(count)
```
